```python
import cv2
import os
import numpy as np

folder_path = r"D:\Training\Pandas\English\Fnt\s(1)"
padding_top = 50

for filename in os.listdir(folder_path):
    image_path = os.path.join(folder_path, filename)
    img = cv2.imread(image_path)

    # Get the height and width of the image
    height, width = img.shape[:2]

    # Create a new blank image with the desired size (width x height + padding_top)
    new_img = np.ones((height + padding_top, width, 3), dtype=np.uint8)

    # Copy the original image to the new image with padding
    new_img[padding_top:, :] = img

    new_path = 'English/Fnt/s(2)/'
    # Save the new image with padding
    new_image_path = os.path.join(folder_path, f"padded_{filename}")
    cv2.imwrite(new_image_path, new_img)
```
